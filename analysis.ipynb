```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

users_df = pd.read_csv('users.csv')
repos_df = pd.read_csv('repositories.csv')

sns.set(style="whitegrid")


```

```python
# 1. Distribution of Followers per User
plt.figure(figsize=(10, 6))
sns.histplot(users_df['followers'], bins=30, kde=True, color="skyblue")
plt.title('Distribution of Followers per User')
plt.xlabel('Number of Followers')
plt.ylabel('Count of Users')
plt.show()
```

```python

# 2. Distribution of Public Repositories per User
plt.figure(figsize=(10, 6))
sns.histplot(users_df['public_repos'], bins=30, kde=True, color="salmon")
plt.title('Distribution of Public Repositories per User')
plt.xlabel('Number of Public Repositories')
plt.ylabel('Count of Users')
plt.show()
```

```python

# 3. Top 10 Companies Represented
top_companies = users_df['company'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_companies.values, y=top_companies.index, palette="viridis")
plt.title('Top 10 Companies Represented by Users')
plt.xlabel('Number of Users')
plt.ylabel('Company')
plt.show()

```

```python
# 4. Most Popular Languages in Repositories
top_languages = repos_df['language'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_languages.values, y=top_languages.index, palette="magma")
plt.title('Top 10 Programming Languages in Repositories')
plt.xlabel('Number of Repositories')
plt.ylabel('Language')
plt.show()

```

```python
# 5. Most Common Licenses in Repositories
top_licenses = repos_df['license_name'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_licenses.values, y=top_licenses.index, palette="coolwarm")
plt.title('Top 10 Licenses in Repositories')
plt.xlabel('Number of Repositories')
plt.ylabel('License Type')
plt.show()

```

```python
# 6. Stars and Watchers Distribution per Repository
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
sns.histplot(repos_df['stargazers_count'], bins=30, kde=True, color="gold", ax=ax[0])
ax[0].set_title('Distribution of Stars per Repository')
ax[0].set_xlabel('Number of Stars')
ax[0].set_ylabel('Count of Repositories')

sns.histplot(repos_df['watchers_count'], bins=30, kde=True, color="purple", ax=ax[1])
ax[1].set_title('Distribution of Watchers per Repository')
ax[1].set_xlabel('Number of Watchers')
ax[1].set_ylabel('Count of Repositories')

plt.tight_layout()
plt.show()
```
